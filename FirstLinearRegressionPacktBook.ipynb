```
# Import libraries

import numpy as np
import matplotlib.pyplot as plot

# Create 1000 points following a function y = .1 * x + .4z
# with some normal random distribution

num_points = 1000
vectors_set = []

for i in range(num_points):
  W = 0.1
  b = 0.4
  x1 = np.random.normal(0.0, 1.0) # inputs are mean and standard deviation
  nd = np.random.normal(0.0, 0.05)
  y1 = W * x1 + b
  
  # Add some impurity with normal distribution
  y1 = y1 + nd
  
  # Append to the vector set
  vectors_set.append([x1, y1])
  
# Separate the data point across axis
x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]

# Plot the data points
plot.plot(x_data, y_data, 'ro', label='Original data')
plot.legend()
plot.show()
```
